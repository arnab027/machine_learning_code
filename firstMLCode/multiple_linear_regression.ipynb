```python
X=startups_df.iloc[:,0:-1]
print(X.values)
y=startups_df.iloc[:,-1]
print(y.values)
```

```python
#apply one hot encoding to State column as the column is not numerical
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
### except the columns for which one hot encoding is implemented other columns will have no change - remainder='passthrough'
### only [3] position column changes will be done
## encoding is OneHotEncoder
ct= ColumnTransformer(transformers= [('encoder', OneHotEncoder(), [3])], remainder='passthrough')
independent_cols_x = np.array(ct.fit_transform(X.values))
print(independent_cols_x)
```
